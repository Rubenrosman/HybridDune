```python
map_in = r'C:\Users\dpoppema\OneDrive - Delft University of Technology\dataset_hybrid_dune\Hydrodynamics\Pressure sensors beach and shoreface (BS)\raw NetCDF\Deployment period 2'

# List all .nc files in the folder
nc_files = [f for f in os.listdir(map_in) if f.endswith('.nc')]

# Update and overwrite attributes
for fname in nc_files:
    nc_in = os.path.join(map_in, fname)
    fname_mod = fname.replace(' P_', '.P:')
    fname2 = fname_mod.replace(' raw data - period 1.nc', '').replace(' raw data - period 2.nc', '')
    with Dataset(nc_in, mode='r+') as ds:
        ds.setncattr('name', fname_mod)
        
        x = ds.summary
        ds.delncattr('summary')
        ds.setncattr('summary', x)
        
        ds.delncattr('contact person')
        ds.setncattr('contact person','Daan Poppema')

        if 'emailadres' in ds.ncattrs(): # if attribute 'emailadres' (Dutch spelling) exists, remove it 
            ds.delncattr('emailadres')
        else:
            ds.delncattr('email address')
        ds.setncattr('email address','d.w.poppema@tudelft.nl')
        
        ds.delncattr('instrument')
        ds.setncattr('instrument', fname2)
        
        x = ds.getncattr('instrument type') # Ruskin RBR Solo
        ds.delncattr('instrument type')
        ds.setncattr('instrument type',x)
        
        x = ds.getncattr('instrument serial number')
        ds.delncattr('instrument serial number')
        ds.setncattr('instrument serial number',x)
        
        x = ds.epsg
        ds.delncattr('epsg')
        ds.setncattr('epsg',x)
        
        x = ds.getncattr('time zone')
        ds.delncattr('time zone')
        ds.setncattr('time zone',x)
        
        x = ds.getncattr('start time')
        ds.delncattr('start time')
        ds.setncattr('start time',x)
        
        x = ds.getncattr('end time')
        ds.delncattr('end time')
        ds.setncattr('end time',x)
        
        x = ds.version
        ds.delncattr('version')
        ds.setncattr('version',x)

        ds.delncattr('modification datetime')  
        ds.setncattr('modification datetime', datetime.now().strftime('%d-%b-%Y %H:%M'))
        if 'construction datetime' in ds.ncattrs():         # if attribute 'construction datetime' exists, remove it
            ds.delncattr('construction datetime')
        
        x = ds.Conventions
        ds.delncattr('Conventions')
        ds.setncattr('Conventions',x)

        x = ds.comment_1
        ds.delncattr('comment_1')
        ds.setncattr('comment_1',x)
        if 'url of online dataset' in ds.ncattrs(): # if attribute 'emailadres' (Dutch spelling) exists, remove it 
            ds.delncattr('url of online dataset')
        print(f"Update: {nc_in}")
ds
```
