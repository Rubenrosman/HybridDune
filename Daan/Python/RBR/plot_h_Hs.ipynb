```python
import numpy as np
import xarray as xr
import os   
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nc_1 = r'O:\HybridDune experiment\data RBR, OSSI\netcdf\processed\Deployment period 1\Pressure sensor S3P3 RBR6 processed data - period 1.nc'
nc_1b = r'O:\HybridDune experiment\data RBR, OSSI\netcdf\processed\Deployment period 1\Pressure sensor S3P2 Ossi2 processed data - period 1.nc'

nc_2 = r'O:\HybridDune experiment\data RBR, OSSI\netcdf\processed\Deployment period 2\Pressure sensor S3P3 RBR6 processed data - period 2.nc'

ds1 = xr.open_dataset(nc_1)
ds1b = xr.open_dataset(nc_1b)
ds2 = xr.open_dataset(nc_2)
```

```python
ds1

```

```python
# Period 1
t = ds1.t_block
h = ds1.h_mean
Hs = ds1.Hm0

h_filter = h.where((h >= 1) )
Hs_filter = Hs.where((h >= 1)) 

t_storm1 = pd.to_datetime('2024-12-19 17:30')
t_storm2 = pd.to_datetime('2024-12-22 07:30')
t_storm3 = pd.to_datetime('2025-01-01 15:00')
t_storm4 = pd.to_datetime('2025-01-03 17:00')
t_storm5= pd.to_datetime('2025-01-06 19:00')

# Create a 3-panel plot: t vs h, t vs Hs, t vs Hs/h
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

# t vs h
axs[0].plot(t, h, color='black',label='h<1m')
axs[0].plot(t, h_filter, label='h>1m')
axs[0].axvline(t_storm1, color='black', linestyle='--')
axs[0].axvline(t_storm2, color='black', linestyle='--')
axs[0].set_ylabel('water depth (m)')
axs[0].legend()
axs[0].grid(True)

# t vs Hs
axs[1].plot(t, Hs, label='Hm0 (h<1m)', color='black')
axs[1].plot(t, Hs_filter, label='Hm0 (h>1m)', color='orange')
axs[1].axvline(t_storm1, color='black', linestyle='--')
axs[1].axvline(t_storm2, color='black', linestyle='--')
axs[1].set_ylabel('Hm0')
axs[1].legend()
axs[1].grid(True)

# t vs Hs/h
axs[2].plot(t, Hs/h, label='Hm0/h (h<1m)', color='black')
axs[2].plot(t, Hs_filter/h_filter, label='Hm0/h (h>1m)', color='green')
axs[2].axvline(t_storm1, color='black', linestyle='--')
axs[2].axvline(t_storm2, color='black', linestyle='--')
axs[2].set_xlabel('Time')
axs[2].set_ylabel('Hm0/h')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
