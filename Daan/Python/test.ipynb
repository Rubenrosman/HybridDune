```python
# 1. Open the file
infile = r"O:\HybridDune experiment\data Kellers\NetCDF\Pressure sensors structure interface - file 9, 8-10jan2025 (after storm 5), S4 Wall-in-dune, 10Hz.nc"
# Automatically create outfile name with _filtered before .nc
base, ext = os.path.splitext(infile)
outfile = base + "_filtered" + ext

with xr.open_dataset(infile, engine="netcdf4", chunks={}) as ds:
    # Drop variable 'temp'
    ds = ds.drop_vars('temp', errors='ignore')    

    # 2. Make voltage[0, :, :] nan
    voltage = ds['voltage'].copy()
    voltage_data = voltage.values
    voltage_data[0, :, :] = np.nan
    voltage.values = voltage_data
    
    # assign to ds_new
    ds_new = ds.copy()
    ds_new['voltage'] = voltage

    # 4. Save with same dtype, chunks, and compression
    encoding = ds.encoding.copy()
    for var in ds_new.data_vars:
        if var in ds.variables and hasattr(ds[var], 'encoding'):
            ds_new[var].encoding = ds[var].encoding.copy()
    ds_new.to_netcdf(outfile, engine="netcdf4")

print(f"Filtered file saved as: {outfile}")
```
