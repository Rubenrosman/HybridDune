```python
# Find all files in folder_in ending with '1000Hz.nc'
folder_in = r'O:\HybridDune experiment\data Kellers\NetCDF'
file_pattern = os.path.join(folder_in, '*1000Hz.nc')
file_list = glob.glob(file_pattern)
file_list
```

```python
# Efficiently process only every 100th timestep at file loading using xarray's open_dataset with isel
for nc_in in file_list:
    nc_smaller = nc_in.replace('1000Hz.nc', '10Hz.nc')
    print(f"Processing {nc_in} -> {nc_smaller}")

    # Open only every 100th timestep using isel
    with xr.open_dataset(nc_in, chunks='auto') as ds_in:
        # Get the indices for every 100th timestep
        t_indices = np.arange(0, ds_in.dims['t'], 100)
        ds_small = ds_in.isel(t=t_indices)
        vars_to_drop = ['temp'] if 'temp' in ds_small.variables else []
        ds_small = ds_small.drop_vars(vars_to_drop)
        ds_small.attrs['timestep'] = '0.1 sec'

        # Compression for saving netcdfs
        encoding = {'voltage': { 'scale_factor': 0.001, 'dtype': 'int16', '_FillValue': -9999, 'shuffle': True, 'chunksizes': (1, 1, 72000)},
                    't': {'chunksizes': (72000,) } }
        compression = {var: {"zlib": True, "complevel": 4} for var in list(ds_small.data_vars) + list(ds_small.coords)}
        for var, comp in compression.items():
            if var in encoding:
                encoding[var].update(comp)
            else:
                encoding[var] = comp
        ds_small.to_netcdf(nc_smaller, encoding=encoding, compute=True)
        print(f"Saved {nc_smaller}")
```
