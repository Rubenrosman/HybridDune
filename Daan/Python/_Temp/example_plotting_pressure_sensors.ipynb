```python
# Example plots of pressure data Hybrid Dune Campaign

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
```

```python
nc_raw = r'U:\HybridDune\,Dataset\Pressure sensors\raw NetCDF\Deployment period 1\Pressure sensor S3P3 RBR6 raw data - period 1.nc'
nc_QC  = r'U:\HybridDune\,Dataset\Pressure sensors\QC\Deployment period 1\Pressure sensor S3P3 RBR6 p_rel - period 1.nc'
nc_processed = r'U:\HybridDune\,Dataset\Pressure sensors\processed\Deployment period 1\Pressure sensor S3P3 RBR6 processed data - period 1.nc'

ds_raw = xr.open_dataset(nc_raw)
ds_QC = xr.open_dataset(nc_QC)
ds_processed = xr.open_dataset(nc_processed)

ds_processed # print the variables in ds_processed
```

```python
# --- Plot raw pressure data ---
fig, ax = plt.subplots(figsize=(10, 5))

# Plot raw pressure (Pa → dbar if you like: ds_raw.p / 1e4)
ax.plot(ds_raw.t, ds_raw.p, label='Raw Pressure (Pa)', color='tab:blue')

# Axes and labels
ax.set_xlabel('Time')
ax.set_ylabel('Pressure [Pa]', color='tab:blue')
ax.tick_params(axis='y', labelcolor='tab:blue')

# Grid, title, legend
ax.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
ax.set_title(f"{ds_raw.attrs.get('instrument', 'Unknown instrument')}, Deployment period 1")
ax.legend(loc='upper right')

fig.tight_layout()
plt.show()
```

```python
# plot QC data

# determine moving mean of ds_QC.p
p = ds_QC.p_rel/10000 # convert to m
window_size = ds_QC.sf.values * 60 * 10  # window size: 16 hz * 60s * 10min
p_moving_mean = p.rolling(t=window_size, center=True).mean()

# --- Plot QC'd relative pressure data ---
fig, ax = plt.subplots(figsize=(10, 5))

# Plot relative pressure and its moving mean
ax.plot(ds_QC.t, p, label='Relative Pressure', color='tab:blue')
ax.plot(ds_QC.t, p_moving_mean, label='Relative Pressure (Moving Mean)', color='orange')

# Axes and labels
ax.set_xlabel('Time')
ax.set_ylabel('Pressure [m]', color='tab:blue')
ax.tick_params(axis='y', labelcolor='tab:blue')

# Grid, title, legend
ax.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
ax.set_title('Relative pressure (QC processed, 10 min avg)')
ax.legend(loc='upper right')

fig.tight_layout()
plt.show()
```

```python
# plot processed data: timeseries
# plt.figure(figsize=(10, 5))
# plt.plot(ds_processed.t_full, ds_processed.zs, label='Water level')
# plt.plot(ds_processed.t_block, ds_processed.zs_mean, label='Mean water level (block avg)', color='orange') # block averaged water level, so per 20 minutes
# plt.axvline(pd.Timestamp('2024-12-19 18:00'), color='red', linestyle='--', label='Peak storm 1') # manually plot line at 19 dec, 18:00

# plt.xlabel('Time')
# plt.ylabel('Water level [m NAP]')
# plt.legend(loc="upper left")


# --- Plot processed water level time series (homogenised style) ---
fig, ax = plt.subplots(figsize=(10, 5))

# Plot full-resolution and block-averaged water level
ax.plot(ds_processed.t_full, ds_processed.zs, 
        label='Water level', color='tab:blue')
ax.plot(ds_processed.t_block, ds_processed.zs_mean, 
        label='Mean water level (block avg)', color='orange')

# Mark peak storm event
ax.axvline(pd.Timestamp('2024-12-19 18:00'), 
           color='red', linestyle='--', label='Peak storm 1')

# Axes and labels
ax.set_xlabel('Time')
ax.set_ylabel('Water level [m NAP]', color='tab:blue')
ax.tick_params(axis='y', labelcolor='tab:blue')

# Grid, title, legend
ax.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
ax.set_title('Processed water level (20 min block averages)')
ax.legend(loc='upper right')

fig.tight_layout()
plt.show()
```

```python
# Plot wave data
plt.figure(figsize=(10, 5))

# Primary axis (left one)
fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(ds_processed.t_block, ds_processed.Hm0, label=r'$H_{m0}$ [m]', color='tab:blue')
# ax1.set_xlim(pd.Timestamp('2024-12-15'), pd.Timestamp('2025-01-15'))
ax1.set_xlabel('Time')
ax1.set_ylabel(r'$H_{m0}$ [m]', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(True, linestyle='--', alpha=0.7)

# Secondary axis (right one)
ax2 = ax1.twinx()
ax2.plot(ds_processed.t_block, ds_processed.Tp, label=r'$T_{p}$ [s]', color='tab:orange', alpha=0.7)
ax2.set_ylabel(r'$T_{p}$ [s]', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Title and legend
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper right')
plt.ylim(0, 20)
plt.tight_layout()
plt.show()


# Plot spectrum
psd = ds_processed.psd.sel(t_block=pd.Timestamp('2024-12-19 18:40'))
frequencies = ds_processed.f

fig, ax = plt.subplots(figsize=(10, 5))

# Plot spectrum
ax.plot(frequencies, psd, color='tab:blue', label='Wave Spectrum')

# Axes and labels
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Spectral density [m²/Hz]', color='tab:blue')
ax.tick_params(axis='y', labelcolor='tab:blue')
ax.set_xlim(0, 1)

# Grid, title, legend
ax.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
time_str = pd.to_datetime(psd.t_block.values).strftime('%d-%m-%Y %H:%M')
ax.set_title(f'Wave spectrum at {time_str}')
ax.legend(loc='upper right')

fig.tight_layout()
plt.show()

```
