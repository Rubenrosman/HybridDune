```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Step 1: Define the file path
file_path = r'C:\Users\ruro\OneDrive - Boskalis\Documents\Msc. Thesis\Lidar_cal\Lidar_cal_20241223\Lidar_1\Lidar1_TS_20241223_11.37.45.txt'

# Step 2: Read the file
data = pd.read_csv(file_path, sep='\t', header=None, skip_blank_lines=True, comment='#', engine='python')
# data = data.iloc[1:]
# Extract necessary columns
names = data[0]  # 1st column: Names
x = data[3]      # 4th column: X coordinates
y = data[4]      # 5th column: Y coordinates
z = data[5]      # 6th column: Z coordinates

# Step 3: Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z)

# Step 4: Add labels to the points
for i, name in enumerate(names):
    ax.text(x[i], y[i], z[i], name, fontsize=8)

# Customize the plot
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.title('3D Scatter Plot with Labels')

# Step 5: Show the plot
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(x, y, color='blue', label='Data Points')

# Add labels for each point
for i, name in enumerate(names):
    plt.text(x[i], y[i], name, fontsize=9, ha='right', va='bottom')  # Adjust alignment as needed

# Customize the plot
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Scatter Plot of X vs Y with Names')
plt.legend()
plt.grid()
plt.show()
```
